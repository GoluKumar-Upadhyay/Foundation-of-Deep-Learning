# Data Inspection & Understanding

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "Aman", "Sonu"],
    "Age": [25, 30, 35, 25, 30],
    "Salary": [50000, 60000, 70000, 50000, 60000],
    "Department": ["IT", "HR", "IT", "IT", "HR"]
}
df = pd.DataFrame(data)
```

```python
print(df.head(2))
```

```python
print(df.tail(2))
```

```python
print(df.shape)
```

```python
df.info()
```

```python
print(df.describe())
```

```python
df["Department"].value_counts()
```

```python
print(df.columns)
```

```python
name_row0 = df.loc[0, "Name"]
print("\nValue at row 0, column 'Name':", name_row0)

subset = df.loc[0:2, ["Name", "Salary"]]
print("\nSubset using loc:\n", subset)
```

```python
val = df.iloc[1, 2]
print("\nValue at 2nd row, 3rd column:", val)

subset = df.iloc[0:3, 0:2]
print("\nSubset using iloc:\n", subset)
```

```python
salary_first = df.at[0, "Salary"]
print("\nSalary of first row using at:", salary_first)

salary_second = df.iat[1, 2]
print("Salary of 2nd row using iat:", salary_second)
```

```python
names = df["Name"]
print("\nSingle column 'Name':\n", names)

subset = df[["Name", "Salary"]]
print("\nMultiple columns 'Name' and 'Salary':\n", subset)
```

# Handling Missing Data

```python
record = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie", "David", None],
    "Age": [25, 30, None, 40, 35],
    "Salary": [50000, None, 70000, None, 60000],
    "Department": ["IT", "HR", None, "Finance", "IT"]
})
print(df)
```

```python
record.isnull()
```

```python
record.notnull()
```

```python
record.dropna()
```

```python
record.fillna(0)
```

```python
record["Salary"] = record["Salary"].interpolate()
record
```

```python
record["Age"] = record["Age"].fillna(record["Age"].mean())
record
```

```python
record["Salary"] = record["Salary"].fillna(record["Salary"].median())
record
```

```python
record["Department"].fillna(record["Department"].mode()[0])
record
```
