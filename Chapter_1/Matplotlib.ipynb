```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["Species"] = iris.target
```

```python
import numpy as np

species_means = df.groupby(iris.target)["petal length (cm)"].mean()
plt.bar(["Setosa", "Versicolor", "Virginica"], species_means)
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()
```

```python
df["species"] = iris.target

species_means = df.groupby("species")["petal length (cm)"].mean()

# Category labels
species_names = ["Setosa", "Versicolor", "Virginica"]

# Advanced Bar Plot
plt.figure(figsize=(8,5))

plt.bar(
    x=species_names,                          # Categories
    height=species_means,                     # Numerical values
    color=['skyblue','lightgreen','salmon'],  # Bar colors
    width=0.6,                                # Bar width
    align='center',                           # Alignment
    edgecolor='black',                        # Border color
    label='Average Petal Length',             # Legend label
    alpha=0.8                                 # Transparency
)
```

```python
plt.scatter(df["sepal length (cm)"], df["sepal width (cm)"])
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()
```

```python
plt.figure(figsize=(8,5))

scatter = plt.scatter(
    x=df["sepal length (cm)"],             # X values
    y=df["petal length (cm)"],             # Y values
    c=iris.target,                         # Color by species
    s=df["sepal width (cm)"] * 40,         # Size based on another feature
    marker='o',                            # Marker shape
    alpha=0.7,                             # Transparency
    edgecolors='black',                    # Border color
    linewidths=0.5,                        # Border thickness
    label="Flowers"                        # Legend label
)

plt.title("Advanced Scatter Plot: Sepal vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True, linestyle='--', alpha=0.4)

plt.legend(*scatter.legend_elements(), title="Species")
plt.tight_layout()
plt.show()
```

```python
plt.plot(df["petal length (cm)"])
plt.title("Line Plot: Petal Length Trend")
plt.xlabel("Index")
plt.ylabel("Petal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.plot(
    df["petal length (cm)"],          # Y values
    color='purple',                   # Line color
    linestyle='--',                   # Dashed line
    linewidth=2.5,                    # Line thickness
    marker='o',                       # Marker shape
    markersize=5,                     # Marker size
    label="Petal Length",             # Legend label
    alpha=0.85                        # Transparency
)

plt.title("Advanced Line Plot: Petal Length Trend")
plt.xlabel("Sample Index")
plt.ylabel("Petal Length (cm)")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
plt.boxplot(df["petal width (cm)"])
plt.title("Box Plot: Petal Width Distribution")
plt.ylabel("Petal Width")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.boxplot(
    df["petal length (cm)"],          # Single feature  one box
    notch=True,                       # Notched box (median confidence)
    vert=True,                        # Vertical boxplot
    patch_artist=True,                # Allows box color fill
    showfliers=True,                  # Show outliers
    boxprops=dict(facecolor='lightgreen', color='black'),  # Box style
    medianprops=dict(color='red', linewidth=2),            # Median line
    whiskerprops=dict(color='blue'),                       # Whiskers
    capprops=dict(color='black')                           # Caps
)

plt.title("Advanced Box Plot of Petal Length")
plt.ylabel("Petal Length (cm)")
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.savefig("box_plot_advanced.png")
plt.show()
```

```python
df["species"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Pie Plot: Species Distribution")
plt.ylabel("")
plt.show()
```

```python
df["species"] = iris.target
species_counts = df["species"].value_counts()

plt.pie(
    species_counts,                                    # Slice sizes
    labels=["Setosa", "Versicolor", "Virginica"],      # Slice labels
    autopct='%1.1f%%',                                 # Show percentage
    startangle=90,                                     # Rotate start angle
    colors=['lightcoral','lightskyblue','lightgreen'], # Slice colors
    explode=(0.05,0.05,0.05),                        # Pull slices out slightly
    shadow=True                                       # Add shadow
)

plt.title("Species Distribution in Iris Dataset")  
plt.show()                                        
```

```python
counts = df["Species"].value_counts()
labels = ["Setosa", "Versicolor", "Virginica"]
colors = ['#FF9999','#66B2FF','#99FF99']
plt.figure(figsize=(5,4))
plt.pie(counts, labels=labels, autopct="%1.1f%%", colors=colors, startangle=90, wedgeprops={'edgecolor':'black'})
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
plt.gca().add_artist(centre_circle)
plt.title("Donut Plot: Species Distribution", fontsize=14)
plt.show()
```

```python
plt.hist(df["sepal width (cm)"], bins=10)
plt.title("Histogram: Sepal Width")
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(
    df["sepal width (cm)"],          # Numeric data
    bins=15,                          # Number of bins
    range=(0,8),                      # Min & Max values
    color='lightgreen',               # Fill color
    edgecolor='black',                # Border color
    alpha=0.7,                        # Transparency
    label='Sepal width',              # Legend
    density=False,                    # Counts (True  probability)
    histtype='stepfilled'             # Filled step style
)

plt.title("Advanced Histogram: Sepal width")
plt.xlabel("Sepal width (cm)")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
```
