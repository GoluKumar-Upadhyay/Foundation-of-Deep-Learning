```python
#Matplotlib------------------------------------------------------------
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
species_means = df.groupby(iris.target)["petal length (cm)"].mean()
plt.bar(["Setosa", "Versicolor", "Virginica"], species_means)
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")


#seaborn---------------------------------------------------------------
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris


iris = sns.load_dataset('iris')  

sns.barplot(x='species', y='petal_length', data=iris)
plt.title("Bar Plot: Species Count")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')  

sns.barplot(x='species', y='petal_length', data=iris)
plt.title("Bar Plot: Species Count")
plt.show()
```

```python
sns.barplot(
    x="species",              # Category column
    y="petal_length",         # Correct numeric column name
    data=iris,
    estimator=np.mean,        # Show mean value
    ci=95,                    # 95% confidence interval
    palette="pastel",         # Bar colors
    capsize=0.2               # Caps on error bars
)

plt.title("Average Petal Length by Species")
plt.show()


```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title("Scatter Plot: Sepal Length vs Sepal Width")
plt.show()
```

```python
iris = sns.load_dataset('iris')

sns.scatterplot(
    x="sepal_length",        # X-axis feature
    y="petal_length",        # Y-axis feature
    hue="species",           # Color by species
    style="species",         # Marker shape by species
    size="sepal_width",      # Bubble size
    sizes=(40, 200),         # Size range
    palette="deep",          # Color palette
    alpha=0.7,               # Transparency
    edgecolor="black",       # Marker border color
    linewidth=0.5,           # Border thickness
    legend="brief",           # Full legend
    data=iris
)

plt.title("Advanced Scatter Plot of Iris Features")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.lineplot(x=range(20), y=iris['sepal_length'][:20], marker='o')
plt.title("Line Plot: Sepal Length (First 20 samples)")
plt.xlabel("Sample Index")
plt.ylabel("Sepal Length")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(
    x=range(20),
    y=iris['sepal_length'][:20],
    marker='o',                       # Circle markers
    markersize=8,                     # Marker size
    linestyle='--',                   # Dashed line
    linewidth=2.5,                    # Thicker line
    color='mediumseagreen',           # Stylish color
    alpha=0.85                        # Slight transparency
)

plt.grid(True, linestyle=':', linewidth=0.8, alpha=0.7)
plt.title("Line Plot: Sepal Length (First 20 Samples)", fontsize=14, fontweight='bold', color='black')
plt.xlabel("Sample Index", fontsize=12, fontweight='bold')
plt.ylabel("Sepal Length (cm)", fontsize=12, fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.title("Box Plot: Sepal Length by Species")
plt.show()
```

```python
sns.boxplot(
    x="species",             # Species on x-axis
    y="petal_length",        # Petal length on y-axis
    palette="Set3",          # Box colors
    width=0.5,               # Box width
    fliersize=5,             # Size of outlier points
    linewidth=1.5,           # Box edge thickness
    data=iris                # DataFrame containing data
)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.title("Boxplot: Petal Length by Species", fontsize=14, fontweight='bold', color='black')
plt.xlabel("Species", fontsize=12, fontweight='bold')
plt.ylabel("Petal Length (cm)", fontsize=12, fontweight='bold')
plt.show()

```

```python
plt.figure(figsize=(6,4))
sns.violinplot(x='species', y='sepal_width', data=iris)
plt.title("Violin Plot: Sepal Width by Species")
plt.show()
```

```python
plt.figure(figsize=(8,5))  # Figure size

sns.violinplot(
    x="species",            # X-axis
    y="sepal_width",        # Y-axis
    palette="muted",        # Colors
    split=False,            # Split violin
    inner="quartile",       # Quartiles
    scale="width",          # Width scale
    data=iris               # Data
)

plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.title("Violin Plot: sepal width by Species") 
plt.xlabel("Species")                             
plt.ylabel("Sepal width")          
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(iris['sepal_length'], bins=15, kde=True)
plt.title("Histogram: Sepal Length Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))  # Figure size

sns.histplot(
    x="sepal_length",  # X-axis
    hue="species",          # Group by species
    bins=15,                # Number of bins
    kde=True,               # Kernel density estimate
    stat="count",           # Show counts
    palette="pastel",       # Colors
    alpha=0.7,              # Transparency
    multiple="stack",       # Stack bars
    data=iris               # Data
)

plt.grid(axis='y', linestyle='--', alpha=0.7)  
plt.title("Histogram: Sepal Length Distribution") 
plt.xlabel("sepal Length ")                
plt.ylabel("Count")                          
plt.show()

```

```python
sns.pairplot(iris, hue='species')
plt.suptitle("Pair Plot: All Feature Relationships", y=1.02)
plt.show()
```

```python
sns.pairplot(
    iris,                    # Data
    hue="species",          # Color by species
    palette="husl",         # Colors
    kind="scatter",         # Scatter plot
    diag_kind="kde",        # KDE on diagonal
    markers=["o","s","D"]   # Marker types for each species
)
plt.show()
```

```python
plt.figure(figsize=(6,4))
iris['sepal_length'][:30].cumsum().plot(kind='area', color='lightgreen', alpha=0.5)
plt.title("Area Plot: Cumulative Sepal Length (First 30 samples)")
plt.xlabel("Sample Index")
plt.ylabel("Cumulative Sepal Length")
plt.show()
```
